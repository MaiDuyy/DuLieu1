1. Kỹ thuật Binning

```python
# 1 Kỹ thuật Binning :Phân tích biến định lượng thanh các khoảng định tính
# Mục đích : 
# - Giảm thiếu dữ liệu nhiêu(noise)
#  - Làm mịn (smooth) dữ liệu
# Hỗ trợ trong nhiều nhiều bài toán Machine Learning như Dcision Tree
```

```python
import pandas as pd

df = pd.DataFrame({'Math': [4, 4, 7, 8, 12, 13, 15, 18, 22, 23, 23, 25],
                   'Eng': [2, 5, 4, 7, 7, 8, 5, 4, 5, 11, 13, 8]})
df.shape
# 12 dòng 2 cột
```

```python
# Chia làm 5 khoảng có số lượng phần tử trong từng khoảng sắp xếp
df['bin1'] = pd.qcut(df['Math'], q=5)
df
```

```python
# Đếm số lượng dữ liệu trong khoảng
df['bin1'].value_counts()
```

```python
# Kỹ thuật chia khoản 
df['bin2'] = pd.qcut(df['Math'], q=[0, .2, .4, .6, .8, 1])
df
```

```python
df['bin2'].value_counts()
```

```python
# Đánh nhãn tưng khoảng
df['bin3'] = pd.qcut(df['Math'],
                           q=[0, .2, .4, .6, .8, 1],
                           labels=['A', 'B', 'C', 'D', 'E'])
df
```

```python
df['bin3'].value_counts()
```

```python
# Phân khoảng phân theo giới hạn 
df['bin4'] = pd.cut(df['Eng'],
                      bins=[0, 5, 8, float('Inf')],
                      labels=['Bad', 'OK', 'Good'])
df
```

```python
df['bin4'].value_counts()
```

1. Lable Encoding (Từ định tính (text), biến phân loại -> thành số )

```python
import pandas as pd

gdp = [11, 22, 33, 44, 55, 66, 77]
countries = ['Spain', 'France', 'Spain', 'Germany', 'France']
# Create data frame from list data
df = pd.DataFrame(list(zip(gdp, countries)),
                  columns=['GDP', 'Countries'])
df
```

```python
from sklearn import preprocessing

# Đánh nhãn
label_encoder = preprocessing.LabelEncoder()
df['labelCountries']= label_encoder.fit_transform(df['Countries']) 
df
```

```python
dfOneHotEncoding = pd.get_dummies(df.Countries, prefix='Country')
print(dfOneHotEncoding.head())
```

```python
# Ghép
df = pd.concat([df, dfOneHotEncoding], axis=1)
df
```

```python
# Encoding to vectorzie
from sklearn.preprocessing import LabelBinarizer
y = LabelBinarizer().fit_transform(df.Countries)
y
```

```python
from sklearn.preprocessing import OneHotEncoder
y = OneHotEncoder().fit_transform(df[['Countries']]).toarray()
print(y)
```

PHƯƠNG PHÁP CHUẨN HÓA DỮ LIỆU (TRANSFORM DATA )
1. Normalization 
2. Standradization 

```python
import pandas as pd

year = [2010,2011,2012,2013,2014,2015]

salaryHCM = [8.5,9.0,10.5,12.0,11.0,10.0]

salaryLA = [5.5,6.0,6.0,5.5,6.0,5.0]
```

```python
df = pd.DataFrame(list(zip(year, salaryHCM,salaryLA)),
                  columns=['Year', 'HCM','LA'])
df
```

```python
# Hướng dẫn max absolute scale về đoạn [-1,1]
#  công thức x
max_la = df['LA'].abs().max()
df['macLA'] = df['LA'] / max_la

df
```

```python
max_la = df['HCM'].abs().max()
df['macHCM'] = df['HCM'] / max_la

df
```

```python

from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)

print(scaled_df)
```

```python

vSc = scaler.transform([[4.5,5.75]])
vSc
```

```python
# Hãy cho biết giá trị lương của HCM và LA trước khi cale là bao nhiêu
# Nếu biết trước giá trị 
vR = scaler.inverse_transform([[0.5,0.8]])
vR
```

```python
# Min mã scale về đoạn [0,1]
min_la = df['LA'].min()
max_la = df['LA'].max()
df['mmcLA'] = (df['LA'] - min_la) / (max_la - min_la)

df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)

print(scaled_df)
```

```python
# Ví dụ về Z-Score
# Nếu một ng có thu nhập 15tr
# ở HCM một ng ở LA thu nhập 12 tr
# thoải mái hơn
df[['HCM', 'LA']].describe()
```

```python
# tính z-score
mean_la = df['LA'].mean()
std_la = df['LA'].std()
df['zLA'] = (df['LA'] - mean_la) / std_la

df
```

```python
# Sinh viên tự tính z-score cho HCM
mean_la = df['HCM'].mean()
std_la = df['HCM'].std()
df['zHCM'] = (df['HCM'] - mean_la) / std_la

df
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])

scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)

print(scaled_df)
```

```python

# Hãy dùng Z-Score để đưa ra kết luận: Nếu một người có thu nhập 15 triệu ở HCM, một người ở LA có thu nhập 12 triệu thì người nào sinh hoạt thoải mái hơn
sv = scaler.transform([[15,12]])
sv
```
