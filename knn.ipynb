```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    roc_curve,
    roc_auc_score
)
```

```python
df = pd.read_excel('data_knn.xlsx', sheet_name="data")
df.head()
```

```python
# bước 2
X = df[['t_i','c_i']].values
y = df[['result']].values

X = X.astype(float)
y = y.astype(float)
```

```python
# bước 3 : chia dữ liệu thành training và testing
# Slpiting data set: 7 mẫu train và 7 mẫu test(tỉ lệ 60:40)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.4, 
    random_state=34
)

# Câu hỏi:
# Cho biết trong tập test có boa nhiêu mẫu dữ liệu
# Hãy cho biết index của những dòng dữ liệu nào được lấy ngẫu nhiên vào tập test
# Yêu cầu: sinh viên cát dữ liệu input và output của tập test vào Excel để quan sát
# Lưu ý test_size phải đọc trên đề và random_state
# INput là X-Train OutPut là Y-Train
# Dùng cái test để đánh giá
# , indices_train, indices_test
```

```python
# Bước 4 Huấn luyện mô hình với k=5
knnModel = KNeighborsClassifier(n_neighbors=5)
knnModel.fit(X_train, y_train.ravel())
```

```python
# Dự báo mô hình trên tập X_test
# 1. Giá trị dự báo là giá trị sinh ra từ input khi đưa vào mô hình 
# 2. Giá trị thực tế là giá trị được thu thập trong tự nhiên và không được tính toán 
# 3. Đánh giá mô hinhg là so sánh thực tế và dự báo
y_test_predicted = knnModel.predict(X_test)
y_test_predicted
```

```python
# Hướng daancx sinh viên tính Accuracy dựa trên file Excel
print("Độ chính xác:", knnModel.score(X_test, y_test))
print("Accuracy score:", accuracy_score(y_test, y_test_predicted))
```

```python
# Câu hỏi: Tập giá trị nhãn dể dự báo là gì
classes = knnModel.classes_
classes
```

```python
# Câu hỏi: Tính xác xuất tiền quyết định (trước khi) quyết định trên tập mẫu
y_test_prob = knnModel.predict_proba(X_test)
y_test_prob
```

```python
# Câu hỏi: Nế thiết lập ngưỡng quyết định là 0.85
# Thì kết quả accuracy thay đổi ra sao
print('perdiction with thresold 0.85 :')
y_pred_test_085 = (knnModel.predict_proba(X_test)[:,1]  >= 0.25).astype
y_pred_test_085
```

```python
cf_matrix = confusion_matrix(y_test, y_test_predicted)
print("\nMa trận nhầm lẫn:")
print(cf_matrix)

# TP+TN / SUM
# TN|FP
# FN|TP
```
