29. Hàm aggregate cho phép áp dụng một hoặc nhiều phép tính thống kê cho các cột được chỉ định
- Cột M1 sẽ tính giá trị nhỏ nhất (min) và lớn nhất (max)
- Cột M2 sẽ tính giá trị trung bình (mean) và tổng (sum)

```python
df.aggregate({'M1':['min','max'],'M2':['mean','sum']})
```

30.  Đếm số lượng hàng trong mỗi nhóm của 'KV'

31. Tính toán các phép thống kê trên M1 theo từng nhóm KV
